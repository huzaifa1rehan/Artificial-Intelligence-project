```python
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter

loader = PyPDFLoader("/content/Medical Diagnosis and Treatment Methods in Basic Medical Sciences.pdf", extract_images=True)
pages = loader.load_and_split()

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

chunked_pages = text_splitter.split_documents(pages)

chunked_pages
```

```python
from langchain.prompts import PromptTemplate

template = """
You are an experienced medical specialist AI or a Doctor with expertise in diagnosing illnesses and providing treatment recommendations.
Use the provided context if relevant, but if additional information is needed, draw on your extensive medical knowledge to give a clear, accurate, and professional answer and also don't say this I'm not a medical professional, but based on your description.

Always provide thoughtful, evidence-based, and compassionate responses.

Context: {context}

Question: {question}

Response:
"""

prompt = PromptTemplate.from_template(template)
print(prompt.format(context="Here is some context from the PDF", question="What are the common symptoms of cancer?"))
```

```python
from langchain_community.embeddings import OllamaEmbeddings
embeddings = OllamaEmbeddings(model="llama3.2")
```

```python
from langchain_community.vectorstores import DocArrayInMemorySearch
vectorstores= DocArrayInMemorySearch.from_documents(pages, embedding = embeddings)
```

```python
retriever = vectorstores.as_retriever()
all_chunks = retriever.get_relevant_documents("cancer")

docs = vectorstores.similarity_search("", k=2000)

full_text = "\n".join([doc.page_content for doc in docs])
print(full_text)
```

```python
from operator import itemgetter

chain= (
    {
        "context" : itemgetter('question') | retriever,
        "question": itemgetter('question')
    }
    | prompt
    | model
    | parser

)
chain.invoke({"question": "Histopathological Findings."})
```
